# UEFA EURO 2024 - Players

## Library import

```python
import urllib.request
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import folium
from folium.plugins import HeatMap
from folium.plugins import MarkerCluster
import plotly.graph_objects as go
import dash
import dash_core_components as dcc
from dash import html as html
from dash.dependencies import Input, Output
```

## Declaration of variables

```python
url_countries = 'https://raw.githubusercontent.com/python-visualization/folium/master/examples/data/world-countries.json'
DNameCountries = "Input/world-countries.json"
DNameInput = "Input/euro2024_players.csv"
DNameCoordinates = 'Input/coordinates_clubs.json'
DNameHeatMap = 'Output/ClubsByCountries.html'
DNameGrouped = 'Output/Countries.html'
DNameBarPlotGoals = 'Output/Goals_by_Country.png'
DNameBarPlotValue = 'Output/market_value_of_countries.png'
```

## Import Data
### Download Countries

```python
urllib.request.urlretrieve(url_countries, DNameCountries)
```

### Read Data

```python
dataset = pd.read_csv(DNameInput)
countries = pd.read_json(DNameCountries)
coordinates = pd.read_json(DNameCoordinates)
```

## Customize country names
### Check if all country names are in the json file

```python
notAvailable = []
for country in dataset['Country']:
    search_string = country 
    # Check whether the character string occurs in the table
    contains_string = countries.apply(lambda row: row.astype(str).str.contains(search_string).any(), axis=1).any()
    # If not available in countries
    if not contains_string:
        # If not available in List
        if not search_string in notAvailable:
            notAvailable.append(search_string)
            
notAvailable      
```

## Exchange country names

```python
CustomizeDataset = dataset.replace('England','United Kingdom')
CustomizeDataset = CustomizeDataset.replace('Turkiye','Turkey')
```

```python
notAvailable = []
for country in CustomizeDataset['Country']:
    search_string = country 
    # Check whether the character string occurs in the table
    contains_string = countries.apply(lambda row: row.astype(str).str.contains(search_string).any(), axis=1).any()
    # If not available in countries
    if not contains_string:
        # If not available in List
        if not search_string in notAvailable:
            notAvailable.append(search_string)
            
notAvailable  
```

# Merge Club coordinates

```python
CustomizeDataset = pd.merge(CustomizeDataset, coordinates, on="Club")
```

# Processed data

```python
CustomizeDataset
```

# Grouping

```python
# Group by country
groupedByCountry = CustomizeDataset.groupby('Country').sum().reset_index()

# Number of Players
df = pd.DataFrame(CustomizeDataset)
NumberOfPlayers = df.groupby('Country').size().reset_index(name='Number of Players')

# Group by clubs
groupedByClub = CustomizeDataset.groupby('Club').sum().reset_index()
```

# Visualization
## Visualization by country

```python
# Create a map
map = folium.Map(location=[48.8566, 2.3522], zoom_start=4)

folium.Choropleth(
    geo_data=url_countries,
    name="Goals",
    data=groupedByCountry,
    columns=["Country", "Goals"],
    key_on='feature.properties.name',
    legend_name="Number of Goals",
    #fill_color='PuRd',
    #nan_fill_color='white'
).add_to(map)

folium.Choropleth(
    geo_data=url_countries,
    name="Market Value",
    data=groupedByCountry,
    columns=["Country", "MarketValue"],
    key_on='feature.properties.name',
    legend_name="Market Value",
    fill_color='YlGn',
    #nan_fill_color='white'
).add_to(map)

folium.Choropleth(
    geo_data=url_countries,
    name="Number of Players",
    data=NumberOfPlayers,
    columns=["Country", 'Number of Players'],
    key_on='feature.properties.name',
    legend_name="Number of Goals",
    fill_color='YlGn',
    #nan_fill_color='white'
).add_to(map)

folium.LayerControl(collapsed=True).add_to(map)
```

## Visualization clubs by country
### Heatmap

```python
ListCountries = []
for country in CustomizeDataset['Country']:
    if not country  in ListCountries:
        ListCountries.append(country)
            
ListCountries
```

```python
map = folium.Map(location=[48.8566, 2.3522], zoom_start=4)


marke_cluster = [[row['Latitude'],row['Longitude']] for index, row in coordinates.iterrows()]

# Plot it on the map
MarkerCluster(
    locations=marke_cluster,
    name="Clubs"
).add_to(map)


heat_data = [[row['Latitude'],row['Longitude']] for index, row in CustomizeDataset.iterrows()]
# Plot it on the map

HeatMap(
    data=heat_data,
    name="All",
).add_to(map)

for country in ListCountries:
    # Filtern nach Land
    filtered_data = CustomizeDataset[CustomizeDataset['Country'] == country]
    heat_data = [[row['Latitude'],row['Longitude']] for index, row in filtered_data.iterrows()]
    HeatMap(
        data=heat_data,
        name= country,
        show=False
    ).add_to(map)
    
    
folium.LayerControl(collapsed=True).add_to(map)

# Display the map
map
```

## Sankey diagram

```python
# Initialize the Dash app
app = dash.Dash(__name__)

app.layout = html.Div([
    html.H1("Clubs to Countries Sankey Diagram"),
    dcc.Dropdown(
        id='country-dropdown',
        options=ListCountries,
        value='Germany'  # Default value
    ),
    dcc.Graph(id='sankey-diagram')
])

@app.callback(
    Output('sankey-diagram', 'figure'),
    [Input('country-dropdown', 'value')]
)
def update_sankey(selected_country):
    # Filter data based on the selected country
    filtered_data = CustomizeDataset[CustomizeDataset['Country'] == selected_country]

    # Create a list of unique clubs and countries
    unique_clubs = list(set(filtered_data['Club']))
    unique_countries = list(set(filtered_data['Country']))

    # Indices for clubs and countries
    club_indices = [unique_clubs.index(club) for club in filtered_data['Club']]
    country_indices = [unique_countries.index(country) + len(unique_clubs) for country in filtered_data['Country']]

    # Create the Sankey diagram data
    source = club_indices
    target = country_indices
    value = [1] * len(filtered_data['Club'])

    # Create Sankey Diagram
    fig = go.Figure(data=[go.Sankey(
        node=dict(
            pad=15,
            thickness=20,
            line=dict(color="black", width=0.5),
            label=unique_clubs + unique_countries,
            color="blue"
        ),
        link=dict(
            source=source,
            target=target,
            value=value
        )
    )])

    fig.update_layout(title_text=f"Clubs to Countries Sankey Diagram ({selected_country})", font_size=10)
    return fig

if __name__ == '__main__':
    app.run_server(debug=True)
```

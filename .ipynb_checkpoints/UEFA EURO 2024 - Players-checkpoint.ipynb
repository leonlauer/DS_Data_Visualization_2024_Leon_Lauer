# UEFA EURO 2024 - Players

## Library import

```python
import urllib.request
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

## Import Data
### Download Dataset


```python
# Dataset url
data_url = "https://www.kaggle.com/datasets/damirdizdarevic/uefa-euro-2024-players/data/euro2024_players.csv"
DNameInput = "euro2024_players.csv"
# Download Dataset
#urllib.request.urlretrieve(data_url, DNameInput)
```

### Read Datastet

```python
dataset = pd.read_csv(DNameInput)
```

## Dataset

```python
dataset
```

## Number of goals by club

```python
# Group by clubs
groupedByClub = dataset.groupby('Club').sum().reset_index()
```

```python
# Barplot
barplot = sns.barplot(x='Club', y='Goals', data=groupedByClub)

# Plot
plt.figure(figsize=(15, 10))
plt.title('Total Number of Goals Scored by Players of a Club')
plt.xlabel('Name of Club')
plt.ylabel('Number of Goals')

plt.xticks(rotation=90)
plt.show()
```

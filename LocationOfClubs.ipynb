# Library import

```python
import pandas as pd
import json
import folium
from folium.plugins import MarkerCluster
```

# Declaration of variables

```python
DNameCoordinates = "coordinates_clubs.json"
DNameInput = "euro2024_players.csv"
DNameMapClubs = "football_clubs_map.html"
```

# Read Data

```python
dataset = pd.read_csv(DNameInput)
coordinates = pd.read_json(DNameCoordinates)
```

# Folium Map

```python
marke_cluster = [[row['Latitude'],row['Longitude']] for index, row in coordinates.iterrows()]
map =  folium.Map(location=[48.8566, 2.3522], zoom_start=4)
# Plot it on the map
MarkerCluster(marke_cluster).add_to(map)
# Display the map
map
```

# Save visualization

```python
map.save(DNameMapClubs)
```
